## Cost function
$$
J(\theta)=\frac{1}{2 m}\left(\sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right)^{2}\right)
$$

```python
def cost(theta, X, y):
    """
    X: R(m*n), m records, n features
    y: R(m)
    theta : R(n), linear regression parameters
    """
    m = X.shape[0]

    inner = X @ theta - y  # R(m*1)

    # 1*m @ m*1 = 1*1 in matrix multiplication
    # but you know numpy didn't do transpose in 1d array, so here is just a
    # vector inner product to itselves
    square_sum = inner.T @ inner #square
    cost = square_sum / (2 * m)

    return cost
```

```python
theta = np.ones(X.shape[1])
cost(theta, X, y)
```

## Gradient
$$
\frac{\partial J(\theta)}{\partial \theta_{0}}=\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)}
$$

```python
def gradient(theta, X, y):
    m = X.shape[0]
    inner = X.T @ (X @ theta - y)  # (m,n).T @ (m, 1) -> (n, 1)
    return inner / m
```

```python
gradient(theta, X, y)
```

## Regular gradient
$$
\frac{\partial J(\theta)}{\partial \theta_{j}}=\left(\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)}\right)+\frac{\lambda}{m} \theta_{j}
$$

```python
def regularized_gradient(theta, X, y, l=1): 
    m = X.shape[0]
    
    regularized_term = theta.copy()  # same shape as theta
    regularized_term[0] = 0  # don't regularize intercept theta (b)
    regularized_term = (l / m) * regularized_term
    
    return gradient(theta, X, y) + regularized_term
```

```python
regularized_gradient(theta, X, y, l=1)
```

## Linear regression

```python
def linear_regression_np(X, y, l=1):
    """linear regression
    args:
        X: feature matrix, (m, n+1) # with incercept x0=1
        y: target vector, (m, )
        l: lambda constant for regularization
    return: trained parameters
    """
    # init theta
    theta = np.ones(X.shape[1])

    # train it
    res = opt.minimize(fun=regularized_cost,
                       x0=theta,
                       args=(X, y, l),
                       method='TNC',
                       jac=regularized_gradient,
                       options={'disp': True})
    return res
```

## Regularized linear regression cost function
$$
J(\theta)=\frac{1}{2 m}\left(\sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right)^{2}\right)+\frac{\lambda}{2 m}\left(\sum_{j=1}^{n} \theta_{j}^{2}\right)
$$

```python
def regularized_cost(theta, X, y, l=1):
    m = X.shape[0]

    regularized_term = (l / (2 * m)) * np.power(theta[1:], 2).sum()

    return cost(theta, X, y) + regularized_term
```

```python
theta = np.ones(X.shape[0])

final_theta = linear_regression_np(X, y, l=0).get('x')
```

```python
b = final_theta[0] # intercept
w = final_theta[1] # slope

plt.scatter(X[:,1], y, label="Training data") 
plt.plot(X[:, 1], X[:, 1]*w + b, label="Prediction")
plt.legend(loc=2)
plt.show()
```

```python
training_cost, cv_cost = [], []
```

## Training steps
1. use the subset of training set to fit the model
2. no regularization when you compute the training cost and CV cost
3. remember to use the same subset of training set to compute training cost

```python
m = X.shape[0]
for i in range(1, m+1):
    res = linear_regression_np(X[:i, :], y[:i], l=0)
    tc = regularized_cost(res.x, X[:i, :], y[:i], l=0)  #test dataset
    cv = regularized_cost(res.x, Xval, yval, l=0)       #cross validation dataset
    
    training_cost.append(tc)
    cv_cost.append(cv)
```

```python
plt.plot(np.arange(1, m+1), training_cost, label='training cost')
plt.plot(np.arange(1, m+1), cv_cost, label='cv cost')
plt.legend(loc=1)
plt.show()
# the model can't even fit training set well, Under Fitting
```

## Create polynomial features

```python
def prepare_poly_data(*args, power):
    """
    args: keep feeding in X, Xval, or Xtest
        will return in the same order
    """
    def prepare(x):
        # expand feature
        df = poly_features(x, power=power)

        # normalization
        ndarr = normalize_feature(df).values

        # add intercept term
        return np.insert(ndarr, 0, np.ones(ndarr.shape[0]), axis=1)

    return [prepare(x) for x in args]
```

```python
def poly_features(x, power, as_ndarray=False):
    data = {'f{}'.format(i): np.power(x, i) for i in range(1, power + 1)} 
    df = pd.DataFrame(data) 

    return df.as_matrix() if as_ndarray else df  
```

```python
X, y, Xval, yval, Xtest, ytest = load_data()
```

```python
poly_features(X, power=3)
```

## prepare polynomial regression data
1. expand feature with power = 8, or any power you want to use 
2. Apply **normalization** to combat $x^n$ situation
3. don't forget intercept term

```python
def normalize_feature(df):
    """Applies function along input axis(default 0) of DataFrame."""
    return df.apply(lambda column: (column - column.mean()) / column.std())  # feature engineering
```

```python
X_poly, Xval_poly, Xtest_poly= prepare_poly_data(X, Xval, Xtest, power=8)
X_poly[:3, :]
```

## Plot learning curve 

```python
def plot_learning_curve(X, y, Xval, yval, l=0):
    training_cost, cv_cost = [], []
    m = X.shape[0]

    for i in range(1, m + 1):
        # regularization applies here for fitting parameters
        res = linear_regression_np(X[:i, :], y[:i], l=l)

        # remember, when you compute the cost here, you are computing
        # non-regularized cost. Regularization is used to fit parameters only
        tc = cost(res.x, X[:i, :], y[:i])
        cv = cost(res.x, Xval, yval)

        training_cost.append(tc)
        cv_cost.append(cv)

    plt.plot(np.arange(1, m + 1), training_cost, label='training cost')
    plt.plot(np.arange(1, m + 1), cv_cost, label='cv cost')
    plt.legend(loc=1)
```

```python
plot_learning_curve(X_poly, y, Xval_poly, yval, l=0) #lamda=0
plt.show()

# the training cost is too low to be true. Over fitting
```

```python
plot_learning_curve(X_poly, y, Xval_poly, yval, l=1) # try lambda=1
plt.show()

# training cost is not 0 anymore
# a little bit over fitting
```

```python
plot_learning_curve(X_poly, y, Xval_poly, yval, l=100)  # try lambda = 100
plt.show()

# too much regularization
# under fitting
```

## find the best $\lambda$

```python
l_candidate = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]
training_cost, cv_cost = [], []
```

```python
for l in l_candidate:
    res = linear_regression_np(X_poly, y, l) 
    
    tc = cost(res.x, X_poly, y)
    cv = cost(res.x, Xval_poly, yval)
    
    training_cost.append(tc) 
    cv_cost.append(cv)
```

```python
plt.plot(l_candidate, training_cost, label='training')
plt.plot(l_candidate, cv_cost, label='cross validation')
plt.legend(loc=2)

plt.xlabel('lambda')

plt.ylabel('cost')
plt.show()
```

```python
# best cv got from all those candidates
l_candidate[np.argmin(cv_cost)]
```

```python
# use test data to check the cost
for l in l_candidate:
    theta = linear_regression_np(X_poly, y, l).x
    print('test cost(l={}) = {}'.format(l, cost(theta, Xtest_poly, ytest)))
```

turns out $\lambda = 0.3$ is even better choice
